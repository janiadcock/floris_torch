```python
plt.rcParams.update({'font.size': 16})

plt.style.use('tableau-colorblind10')                                                          
colors = plt.rcParams['axes.prop_cycle'].by_key()['color']

plt.rcParams.update({'font.size': 20})
plt.rc('lines', linewidth=3.0)

wd = np.array([0, 5, 10, 35, 40, 45, 45.1, 50, 55, 80, 85, 90])
wd_l = np.array([0, 5, 10, 35, 40, 45])
wd_r = np.array([45, 50, 55, 80, 85, 90])
yaw0 = np.zeros(12)
yaw0[:2] = [25, 18.75]
yaw1_l = np.array([25, 18.75, 0, 0, -12.5, -23.4375])
yaw1_r = np.array([23.4375, 12.5, 0, 0, -18.75, -25])
yaw2 = np.zeros(12)
yaw3 = np.zeros(12)
yaw3[-2:] = [-18.75, -25]

plt.plot(wd, yaw0, linestyle='-', color=colors[0], label=str(0))
plt.plot(wd_l, yaw1_l, linestyle='--', color=colors[1], label=str(1))
plt.plot(wd_r, yaw1_r, linestyle='--', color=colors[1]) #label="turbine "+str(1
plt.plot(wd, yaw2, linestyle=':', color=colors[2], label=str(2))
plt.plot(wd, yaw3, linestyle='-.', color=colors[3], label=str(3))

plt.xlabel(r'wd $(^{\circ})$')
plt.ylabel(r'yaw $(^{\circ})$')
plt.xticks([0, 45, 90])
plt.yticks([-25, 0, 25])
plt.legend()
fig = plt.gcf()
fig.set_size_inches(6,6)
```
